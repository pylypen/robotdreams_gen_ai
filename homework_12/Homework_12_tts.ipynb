```python
import os
import csv
import json
from TTS.api import TTS

# Вибір TTS-моделі (приклад для англійської мови)

OUTPUT_DIR = "output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def generate_single_sentence(sentence, filename="simple_output.wav", speaker='tts_models/en/ljspeech/tacotron2-DDC'):
    tts = TTS(model_name=speaker)
    path = os.path.join(OUTPUT_DIR, filename)
    tts.tts_to_file(text=sentence, file_path=path)
    print(f"Збережено: {path}")

def generate_from_csv(file_path):
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            filename = f"{row['id']}.wav"
            generate_single_sentence(row['text'], filename)

def generate_from_json(file_path):
    with open(file_path, encoding='utf-8') as f:
        data = json.load(f)
        for entry in data:
            filename = f"generated{entry['id']}.wav"
            generate_single_sentence(entry['text'], filename)
```

```python
print("=== Демонстрація TTS ===")

# 1. Простий варіант
generate_single_sentence("This is a simple sentence for TTS.")

# 2. Деталізований варіант
generate_from_json("sentences.json")
```

```python
# 4. Альтернативний голос (якщо доступно)
sentence = "Neon lights flickered above rain-soaked streets as drones scanned every movement. Augmented humans whispered secrets through encrypted neural links. In the shadows, a rogue hacker plotted to dismantle the corporate AI that ruled the city’s surveillance grid. Resistance wasn’t just a word—it was survival in a world wired for control."
generate_single_sentence(sentence, "alt_voice.wav", speaker="tts_models/en/jenny/jenny")
```
